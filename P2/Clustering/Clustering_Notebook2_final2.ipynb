**Criterio de Información Bayesiano**

Una forma de discriminar entre posibles agrupamientos, es calcular el Criterio de Información Bayesiano, cuya expresión genérica es:
$$BIC = log L(Datos\mid \hat{\theta}) - \frac{d}{2}log(N)$$
donde $d$ es el número de parámetros libres y $N$ el número de observaciones. A medida que aumenta el número de parámetros la verosimilitud aumenta, pero esto es moderado por el término en que depende de $d$.

Mclust calcula automáticamente el BIC para distintas configuraciones de las matrices covarianza de cada Normal multivariada (que corresponden a distintos números de parámetros) así como distinto número de grupos. Por parsimonia, es recomendable escoger el primer número de grupos que representa un máximo local.

1. Purity

definición: Esta métrica evalúa qué tan "puro" es un clúster identificando cuál es la etiqueta o clase mayoritaria dentro de él. Se basa en calcular la proporción de los objetos que pertenecen a esa clase predominante en comparación con el total de objetos del clúster.

interpretación: Los resultados van de cero (malo) a uno (excelente).

Limitación: Su mayor problema es que no penaliza la creación de demasiados clústeres. Podrías obtener un puntaje perfecto de manera engañosa simplemente poniendo cada objeto en su propio clúster individual.

2. Rand Index

definición: Sirve para comparar dos agrupaciones diferentes (por ejemplo, los clústeres que generó tu modelo frente a una agrupación ideal de referencia). En lugar de mirar clústeres enteros, analiza todas las parejas posibles de datos.

interpretación: Se entiende fácilmente como el porcentaje o fracción de decisiones de agrupación que fueron correctas. Para lograrlo, suma los aciertos (las parejas que correctamente se pusieron en el mismo grupo más las parejas que correctamente se separaron) y los evalúa frente al total de combinaciones posibles. Un detalle clave es que le da exactamente el mismo peso a equivocarse uniendo datos distintos (falsos positivos) que a equivocarse separando datos similares (falsos negativos).

3. Mutual Information

definición: Mide cuánta información comparten dos agrupaciones distintas. Lo hace calculando la probabilidad de que, si eliges un objeto al azar, este pertenezca a un clúster específico en la primera agrupación y a otro clúster específico en la segunda.

limitación y Solución: Tiene una debilidad importante: el puntaje máximo se puede alcanzar de forma artificial si el modelo crea muchísimos clústeres pequeñitos. Para compensar este defecto, normalmente se aplica una versión "normalizada". Esta normalización ajusta el cálculo para que el resultado final siempre sea un valor justo entre cero y uno

```r
# 2. CARGA Y PREPARACIÓN DE DATOS
# Leer el archivo respetando los encabezados y espacios
df <- read.table("..//data//semillas.txt", header = TRUE)

# Separar las variables predictoras (V1 a V6) de la etiqueta real (V7)
X <- df[, 1:6]
y_true <- as.factor(df$V7)

# Escalar (estandarizar) los datos numéricos (media 0, varianza 1)
X_scaled <- scale(X)

# 3. ENTRENAMIENTO DE LOS MODELOS

# Modelo KMeans
set.seed(42) # Semilla para reproducibilidad
kmeans_model <- kmeans(X_scaled, centers = 3, nstart = 25)
y_kmeans <- kmeans_model$cluster

# Modelo GMM (Gaussian Mixture Model)
# El paquete mclust determina internamente las covarianzas. G=3 indica 3 clústeres.
set.seed(42)
gmm_model <- Mclust(X_scaled, G = 3, verbose = FALSE) 
y_gmm <- gmm_model$classification

# 4. FUNCIÓN PARA CALCULAR LA PUREZA (Purity)
# R no tiene una función nativa para Pureza, por lo que la construimos:
purity_score <- function(y_true, y_pred) {
  # Crear tabla de contingencia
  contingency_table <- table(y_true, y_pred)
  # Sumar la cantidad del elemento mayoritario en cada clúster y dividir por el total
  sum(apply(contingency_table, 2, max)) / length(y_true)
}

# 5. CÁLCULO E IMPRESIÓN DE LAS MÉTRICAS

cat("--- Resultados KMeans ---\n")
cat("Pureza (Purity):", round(purity_score(y_true, y_kmeans), 4), "\n")
cat("Índice de Rand Ajustado (ARI):", round(adjustedRandIndex(y_true, y_kmeans), 4), "\n")
cat("Información Mutua Normalizada (NMI):", round(NMI(y_true, y_kmeans), 4), "\n\n")

cat("--- Resultados GMM ---\n")
cat("Pureza (Purity):", round(purity_score(y_true, y_gmm), 4), "\n")
cat("Índice de Rand Ajustado (ARI):", round(adjustedRandIndex(y_true, y_gmm), 4), "\n")
cat("Información Mutua Normalizada (NMI):", round(NMI(y_true, y_gmm), 4), "\n")
```
